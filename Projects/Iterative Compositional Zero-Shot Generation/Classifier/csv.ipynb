```python
import os
import csv
import random

# 定義資料夾名稱對應的屬性和標籤
attribute_label_mapping = {
        'Brown_Hair': 0,
        'Blond_Hair': 1,
        'Gray_Hair': 2,
        'Black_Hair': 3
}

def generate_attr_meta_csv(root_folder, output_csv, train_ratio=0.8):
    # 開啟並寫入csv檔案
    with open(output_csv, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        # 寫入標題行
        writer.writerow(['img', 'category', 'settype'])

        # 遍歷根目錄下的所有資料夾
        for folder_name in os.listdir(root_folder):
            folder_path = os.path.join(root_folder, folder_name)
            print(folder_path)
            # 確保是資料夾
            if os.path.isdir(folder_path):
                # 判斷資料夾名稱是否在 mapping 裡面
                for attr_name, label in attribute_label_mapping.items():
                    if attr_name in folder_name:
                        # 獲取資料夾中的所有圖片
                        all_files = [file_name for file_name in os.listdir(folder_path) if file_name.endswith(('.jpg', '.png', '.jpeg'))]
                        
                        # 打亂順序
                        random.shuffle(all_files)
                        
                        # 計算訓練集和驗證集的分割點
                        train_count = int(len(all_files) * train_ratio)
                        
                        # 分配圖片到train和val
                        train_files = all_files[:train_count]
                        val_files = all_files[train_count:]
                        
                        # 寫入train資料
                        for file_name in train_files:
                            writer.writerow([os.path.join(folder_name, file_name), attr_name, 'train'])
                        
                        # 寫入val資料
                        for file_name in val_files:
                            writer.writerow([os.path.join(folder_name, file_name), attr_name, 'val'])



root_folder='C:/Users/Code/DiT/Classifier/CelebA'
output_csv = 'attr_meta.csv'  # 要輸出的 csv 檔名
# 生成 attr_meta.csv
generate_attr_meta_csv(root_folder, output_csv)
```
